```python
# standardisation of numerical variables
from sklearn.preprocessing import StandardScaler
import pandas as pd
ss=StandardScaler()
# selecting only the columns to standardize
cont=['Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']
# Fit and transform the data
df[cont]=ss.fit_transform(df[cont])
df           #---> Now df will have the standardized values for these columns
```

```python
# Dummy variables
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Initialize One HotEncoder
OHE = OneHotEncoder(sparse_output=False, drop='first')  # Use sparse=False to get a dense array, drop='first' to avoid multicollinerd

# select categorical columns
ctgrcl = ['SKU', 'BU', 'Brand', 'Model']

# Apply OneHotEncoder
encoded_data = OHE.fit_transform(df[ctgrcl])

# Convert the encoded data to a DataFrame
encoded_df = pd.DataFrame(encoded_data, columns=OHE.get_feature_names_out(ctgrcl))

# Drop the original categorical columns and concatenate the encoded columns
df = df.drop(ctgrcl, axis=1)
df = pd.concat([df, encoded_df], axis=1)

# show the result
df
```
